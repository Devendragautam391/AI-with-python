```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace (0,7,100)
y1 = 2*x+1
y2 = 2*x+2
y3 = 2*x+3

plt.title('Line Graphs')
plt.xlabel('Age')
plt.ylabel('Height')

plt.plot(x,y1,"b--",x,y2,"r:",x,y3,"g-.");
plt.legend(['2x+1','2x+2','2x+3'])
```

```python
import numpy as np
import matplotlib.pyplot as plt

x: np.ndarray = np.arange(1,10)
y: np.ndarray = np.array([-0.57,-2.57,-4.80,-7.36,-8.78,-10.52,-12.85,-14.69,-16.78])

plt.scatter(x,y, marker = '+', s=100)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot of (x,y) Points')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("weight-height.csv")

print(df[['Height','Weight']].head(10))
print(df[['Height','Weight']].tail(10))

def create_sample_data(n_samples: int = 10000) -> pd.DataFrame:
    df = pd.DataFrame({
        'Height': np.random.randint(1,100, n_samples),
        'Weight': np.random.randint(1,100, n_samples),
        'id': range(1, n_samples + 1)
    })
    
    df['Height_cm'] = df['Height']*2.54
    df['Weight_kg'] = df['Weight']*0.453592
    return df
    
df = create_sample_data()
print(df[['Height','Height_cm','Weight','Weight_kg']].head(10))
print(df[['Height','Height_cm','Weight','Weight_kg']].tail(10))

print("Mean of Height :", np.mean(df['Height_cm']))
print("Mean of Weight :", np.mean(df['Weight_kg']))
    
plt.hist(df['Height_cm'], bins=30)
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.title('Histrogram of Heights')
plt.show()



```

```python
import numpy as np

A = np.array([[1,2,3],
              [0,1,4],
              [5,6,0]])

A_inv = np.linalg.inv(A)
print("Original matrix A:")
print(A)
print("\nInverse matrix A^-1:")
print(A_inv)

verification = np.dot(A,A_inv)
print("\nVerification (A*A^-1):")
print(verification)
```
